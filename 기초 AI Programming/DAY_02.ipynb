## 리스트와 반복문

### 리스트

여러 개의 데이터를 **한 번에 저장** 가능

````python
# 리스트와 반복문 없이 직접 문자열을 출력하는 경우

print("철수야, 안녕!")
print("영희야, 안녕!")
print("민수야, 안녕!")


# 친구가 3명밖에 없는 경우에는 이렇게 써도 괜찮음.
# 하지만 만약 친구가 100명 이상이라면 이 방법은 비효율적.
# for문과 리스트를 사용하면 코드를 단순화할 수 있음.
````

#### 친구 이름 출력 예제 설명 

1) **리스트**
- 리스트는 **여러 개의 데이터를 한 번에 저장하는 상자**이다.
- [](대괄호) 안에 값을 나열해서 만듦. -> 제일 앞에 있는 값이 0번째
- 각 값은 ,(쉼표)로 구분.

2) **for 반복문**
- 리스트에 있는 친구 이름을 **하나씩 꺼내서** 반복하는 구조

3) **friend**
- friend는 반복할 때마다 **꺼낸 친구 이름을 잠깐 저장하는 공간**
- **반복자**라고 부름.
- 이 변수의 이름은 꼭 friend가 아니어도 됨. (name, person 같은 이름도 가능)

4) **in friends**
- **friends 리스트 안에 있는 값들을 하나씩 꺼내면서 반복한다**는 뜻.

5) **print()**
- print()는 **화면에 내용을 보여주는 기능**

6) **f"{friend}야, 안녕!"**
- **f-string**이라는 문법은 f를 앞에 붙이면 {}(중괄호) 안에 있는 변수를 실제값으로 바꿔주는 것.
- 친구 이름마다 다른 문장을 자동으로 만들어주기 가능.

````python
# 친구 이름 출력 예제

# 친구들의 이름을 리스트에 저장합니다.
friends = ["철수", "영희", "민수"]

# 리스트에 있는 친구들을 한 명씩 꺼내서 인사합니다.
for friend in friends:  # friends 리스트에서 친구 이름을 하나씩 가져옵니다.
    print(f"{friend}야, 안녕!")  # 화면에 "철수야, 안녕!"처럼 출력합니다.
````

````python
# 리스트와 반복문을 활용한 평균 구하기

# 5명의 학생 이름과 점수를 리스트로 준비
students = ["철수", "영희", "민수", "지수", "준호"]
scores = [85, 92, 78, 88, 90]

# 점수들의 총합을 저장할 변수 (초기값은 0)
total = 0

# 총합 설명 문자열 (계산 과정 보여주기 위한 텍스트)
sum_text = "총합= "

# [enumerate() 설명]
# enumerate(scores)를 사용하면 점수 리스트에서 점수를 하나씩 꺼내면서
# '몇 번째 점수인지'를 나타내는 위치번호(index)도 함께 가져올 수 있습니다.

# 예를 들어 scores가 [85, 92]일 때:
# for idx, score in enumerate(scores)를 실행하면:
#   1회차: idx=0, scores=85 (철수 점수)
#   2회차: idx=0, scores=92 (영희 점수)
# 즉, 몇 번째 점수인지와 점수를 동시에 꺼내서 활용할 수 있습니다.

# [len() 설명]
# len(scores)는 '리스트 scores의 길이'를 의미합니다.
# 즉, 점수가 몇 개인지를 알려주는 함수입니다.
# 지금은 학생이 5명이므로 len(scores)는 5입니다.

# 평균을 구할 때 '총합 / 학생 수'에서 바로 len(scores)를 사용합니다.

# 점수 하나씩 꺼내면서 합계 구하고, 각 학생의 이름+점수를 출력용 문자열에 추가
for idx, score in enumerate(scores):
    total += score  # 현재 점수를 총합에 더하기

    # sum_text에 학생 이름과 점수를 추가 (studens[idx]로 이름 가져오기)
    sum_text += f"{students[idx]}({score})"

    # 마지막 학생이 아니면 "+"를 붙여서 다음 학생과 이어줌
    if idx < len(scores) - 1:
        sum_text += " + "

# 평균 계산 (총합 / 학생 수)
average = total / len(scores)

# 최종 결과 출력
print(sum_text + f" = {total}점")  # 총합 설명 출력
print(f"평균 = {total} / {len(scores)}명 = {average:.1f}점")  # 평균 설명 출력
````

## while 반복문

- **조건이 참인 동안 계속 실행되는**, **조건이 맞을 때까지 계속 반복**하는 도구
- for 문이 정해진 횟수만큼 반복한다면, while문은 특정 조건이 만족될 때까지 반복.

### while 기본 문법

```python
while 조건:
    반복할 내용
```
- **while**: 반복을 시작하는 키워드
- **조건**: 반복을 계속할지 멈출지 결정하는 부분
- **반복할 내용**: 조건이 참(True)이면 실행되는 부분

#### 예제: 물을 100도까지 끓이기 설명

1) **물 온도 준비**
- **water_temp**: 물의 온도를 저장

2) **while문 - 물 온도 확인 & 반복**
- 조건을 확인하고, 그 조건이 맞으면 계속 반복

3) **현재 온도 보여주기**
- **f-string**을 사용해서 현재 물 온도 숫자를 넣어서 출력

4) **물 온도 올리기**
- water_temp += 5 : 현재 온도에 5도를 더함.

5) **물이 100도가 되면 반복 종료**
- 물이 100도가 되면 while 반복문이 끝남.

````python
# while문 - 조건이 만족될 때까지 반복하는 반복문
# 예제: 물을 100도까지 끓이기

# 물의 초기 온도를 설정합니다.
water_temp = 80  # 처음 물 온도는 80도입니다.

# 물이 100도보다 낮을 동안 계속 반복합니다.
while water_temp < 100:
    # 현재 물 온도를 보여주고, 물을 더 끓인다고 출력합니다.
    print(f"현재 물 온도는 {water_temp}도입니다. 물을 더 끓입니다.")

    # 물 온도를 5도씩 올립니다.
    water_temp += 5

# 물이 100도에 도달하면 반복문을 빠져나오고 아래 문장이 실행됩니다.
print("물이 100도가 되어 끓이기를 멈춥니다.")
````

#### 비밀번호 비교 프로그램

- **ATM 비밀번호 입력 기능**을 흉내낸 프로그램
- 비밀번호를 **3번까지** 입력할 기회 있음.
- 비밀번호를 맞추면 "서비스 시작"이라고 하고 프로그램이 끝남.
- **3번 다 틀리면** "서비스 이용 제한" 메시지를 띄우고 프로그램이 끝남.

````python
# 정답 비밀번호를 '1234'로 설정
correct_password = "1234"

# 비밀번호 입력 시도 횟수를 0으로 시작
tries = 0

# 비밀번호를 3번까지 입력할 수 있는 반복문 (while)
while tries < 3:
    # 사용자에게 비밀번호 입력 받기
    input_password = input("비밀번호를 입력하세요: ")

    # 입력한 비밀번호가 정답 비밀번호와 같은지 확인
    if input_password == correct_password:
        print("비밀번호 확인 완료! 서비스 시작합니다.")

        # 정답을 맞춘 경우, 더이상 반복할 필요가 없음
        # 그래서 바로 반복문을 탈출 (break 사용)
        break  # <<< 이게 바로 break 역할 !

    else:
        # 비밀번호가 틀린 경우, 시도 횟수 증가
        tries += 1

        # 남은 기회 안내 메시지 출력
        print(f"비밀번호가 틀렸습니다. 남은 기회: {3 - tries}번")

# 반복문이 끝난 후 - 3번 다 틀렸을 경우 안내 메시지 출력
if tries == 3:
    print("비밀번호 3회 실패! 서비스 이용이 제한됩니다.")
````

### 매개변수

- 함수에 데이터를 전달하는 통로, 함수의 재료
- 함수가 작업을 수행하는 데 필요한 정보를 외부에서 받아들이는 변수
- 함수를 더 유연하게 사용할 수 있도록 함.

#### 매개변수의 구조

```python
def 함수이름(재료1, 재료2, ...):
    함수 안에서 할 일
```
- def: "함수를 만들겠다"는 뜻
- 함수이름: 이 함수의 이름
- 재료: 함수가 필요로 하는 입력값(매개변수)
- 함수 안에서 할 일: 실제로 수행할 작업(코드)

````python
# 매개변수 예제: 다양한 라면 끓이기

# 1. 라면 끓이는 함수 만들기 (재료 3개 필요)
def cook_ramen(water, ramen, soup):
    """
    이 함수는 물, 라면 종류, 스프 종류를 받아서
    라면 끓이는 과정을 출력하는 함수입니다.
    """
    print(f"{water}ml 물을 끓입니다.")  # 받은 물로 물 끓이기
    print(f"{ramen}을 넣습니다.")       # 받은 라면 넣기
    print(f"{soup}을 넣습니다.")        # 받은 스프 넣기
    print("라면 완성! 🍜")             # 완성 메시지 출력
    
# 2. 함수 호출 - 재료를 넣어서 라면 끓이기
cook_ramen(500, "신라면", "매운맛 스프")  # 신라면 끓이기
cook_ramen(600, "진라면", "순한맛 스프")  # 진라면 끓이기
````

### 리턴

- 함수가 작업을 완료한 후 결과값을 돌려주는 기능
- 함수의 출력으로, 계산된 결과를 함수 외부로 전달하여 다른 코드에서 활용할 수 있게 한다.

#### 리턴이 필요한 이유
- 리턴이 있으면, **하나의 함수 결과를 다음 함수로 넘길 수 있다.**
- 서로 다른 함수들이 자연스럽게 이어지면서 하나의 작업을 완성할 수 있다.
- 이런 방식이 프로그램을 더 효율적이고 깔끔하게 만들어준다.

#### 리턴의 기본 구조
```python
def 함수이름(재료):
    요리하는 과정 (함수 안에서 할 일)
    return 완성된 요리 (리턴 값)
```
- def: 함수를 만든다는 뜻
- 함수이름: 이 함수의 이름
- 재료: 요리에 필요한 재료(입력값)
- 함수 안에서 할 일: 요리하는 과정
- return: 완성된 요리를 밖으로 보내는 명령어

````python
# 리턴 예제: 두개 공장에서 라면 만들기

# 1. 면을 삶는 함수
def boil_noodle():
    """
    면을 삶는 과정입니다.
    삶아진 면을 결과로 돌려줍니다.
    """
    noodle = "삶아진 면"  # 면을 삶았다고 가정합니다.
    return noodle  # 삶아진 면을 리턴해서 함수 밖으로 보냅니다.

# 2. 삶아진 면을 받아서 라면을 완성하는 함수입니다.
def make_ramen(noodle):
    """
    삶아진 면을 받아서
    스프를 넣고 끓여서 라면을 완성하는 과정입니다.
    완성 메시지를 리턴합니다.
    """
    print(f'{noodle}에 스프를 넣고 끓입니다.')  # 받은 면과 함께 스프 넣기
    message = "라면 완성! 🍜"  # 완성 메시지
    return message  # 완성 메시지를 리턴합니다.

# 3. 함수 호출 순서입니다.
# 먼저 면을 삶고, 삶아진 면을 make_ramen에 전달합니다.

# boil_noodle()을 호출해서 면을 삶고,
# 그 결과를 boiled_noodle이라는 변수에 저장합니다.
boiled_noodle = boil_noodle()

# make_ramen()을 호출하면서, 삶아진 면을 넘기고, 완성 메시지를 돌려받습니다.
complete_message = make_ramen(boiled_noodle)

# 받은 완성 메시지를 화면에 출력합니다.
print(complete_message)
````

- 재사용성: 필요한 곳에서 결과 재활용 가능
- 함수끼리 연결: 결과를 주고받으며 연결 가능

````python
# 숫자 계산 예제

# 1. 두 숫자를 더하는 함수
def add_numbers(a, b):
    """
    두 숫자를 더한 결과를 리턴합니다.
    """
    return a + b

# 2. 받은 숫자를 제곱하는 함수
def square_number(number):
    """
    받은 숫자를 제곱한 결과를 리턴합니다.
    """
    return number * number  # number ** 2 해도 같음

# 3. 함수끼리 연결해서 최종 결과 만들기
sum_result = add_numbers(5, 6)  # 5와 6을 더합니다.
final_result = square_number(sum_result)  # 더한 결과를 제곱합니다.

# 4. 최종 결과 출력
print(f'5와 6을 더한 후 제곱한 결과: {final_result}')
````

### 기본 매개변수

- 함수를 만들 때, **매개변수에 기본값을 미리 정해놓을 수 있음.**
- **그 값을 따로 안 주면 기본값을 사용**하고, **새로운 값을 주면 새로운 값으로 바꿔서 사용**한다.
- 코드의 유연성을 높이고 함수 호출 시 반복되는 값 입력을 줄여준다.
- **선택적 매개변수**가 많은 복잡한 함수에 매우 효과적.

#### 기본 매개변수 구조

```python
def 함수이름(재료1=기본값1, 재료2=기본값2):
    함수 안에서 할 일
```
- 재료 = 기본값처럼 기본값을 설정할 수 있음.
- 기본값이 설정된 매개변수는 함수를 부를 때 생략 가능
- 기본값이 없는 매개변수는 꼭 값을 줘야 함.

````python
# 기본 매개변수 예제: 컵라면 끓이기

def cook_ramen(water=500, ramen='신라면', soup='매운맛 스프'):
    """
    물, 라면 종류, 스프 종류를 받아서 라면 끓이는 과정입니다.
    기본값: 물 500ml, 라면은 신라면, 스프는 매운맛 스프입니다.
    """
    print(f'{water}ml 물을 끓입니다.')  # 물 끓이기
    print(f'{ramen}을 넣습니다.')      # 라면 넣기
    print(f'{soup}을 넣습니다.')       # 스프 넣기
    print("라면 완성! 🍜")            # 완성 메시지 출력

# 기본값으로 호출
cook_ramen()

# 물만 변경
cook_ramen(600)

# 물과 라면 변경
cook_ramen(600, "진라면")

# 물, 라면, 스프 모두 변경
cook_ramen(600, "진라면", "순한맛 스프")

# >>> 기본값 덕분에 다양한 호출 가능
````

### 랜덤 유머 예제

#### API

**ex) 랜덤 유머 예제**
- 이름: JokeAPI
- 역할: 전 세계 유머 문장을 랜덤으로 제공
- URL: https://v2.jokeapi.dev/joke/카테고리
- 지원 카테고리: Any(아무 유머), Programming(프로그래밍 유머) 등
- 언어: 영어, 독일어, 프랑스어 등 (한국어는 미지원)

**URL 설명**
- 기본 주소: https://v2.jokeapi.dev -- API 기본 위치
- 경로: /joke/Programming -- 프로그래밍 유머 요청

#### requests 객체

#### 랜덤 유머 예제 구조

1. API 요청 - 유머집 가게에 전화하기
2. 응답 받기 - 유머집 사장님이 유머를 문자로 보내줌
3. 번역 - 받은 문자를 한국어로 번역
4. 출력 - 번역된 유머를 화면에 보여줌

````python
# 랜덤 유머 예제 구현
import requests
from googletrans import Translator

# googletrans 라이브러리 설치 명령어
# !pip install googletrans==4.0.0-rc1
# 파이썬에서는 번역 기능을 직접 제공하지 않기 때문에 외부 라이브러리 설치해야 함.


# 1. 유머 API에서 유머 문장을 가져오는 함수입니다.
def fetch_joke_from_api(category="Any"):
    """
    유머 문장을 API에서 가져오는 값입니다.
    기본값으로 다음과 같이 설정되어 있습니다.
    - 카테고리: Any (모든 종류의 유머를 가져옵니다.)
    - 언어: en (영어로 된 유머를 가져옵니다.)
    """

    # 유머를 요청할 주소(URL)를 만듭니다.
    # {category} 부분에는 전달받은 카테고리가 들어갑니다.
    # ?lang=en은 영어 유머를 요청한다는 의미입니다.
    url = f"https://v2.jokeapi.dev/joke/{category}?lang=en"

    # 해당 주소로 실제 요청을 보내서 응답을 받습니다.
    response = requests.get(url)

    # 응답으로 받은 내용을 JSON 형식(키와 값으로 구성된 데이터)으로 변환합니다.
    joke_data = response.json()
    
    # API 요청이 실패했는지 확인합니다. (예: 'error' 키 확인)
    if joke_data.get('error'):
        return "유머를 가져오는 데 실패했습니다."

    # 응답으로 받은 데이터에서 유머 종류(type)를 확인합니다.
    # *** 딕셔너리 접근은 대괄호 []를 사용해야 합니다. (수정됨) ***
    if joke_data["type"] == "single":
        # "joke"라는 키에 담긴 한 줄짜리 유머 문장을 꺼내서 저장합니다.
        joke_text = joke_data.get("joke")
    else:
        # "twopart"일 경우, 문답형 유머입니다.
        # "setup"은 질문, "delivery"는 답변입니다.
        joke_text = f"{joke_data.get('setup')} - {joke_data.get('delivery')}"

    # 최종적으로 유머 문장을 함수 결과를 돌려줍니다.
    return joke_text


# 2. 유머를 한국어로 번역하는 함수입니다.
def translate_joke_to_korean(joke_text, lang="ko"):
    """
    영어로 된 유머 문장을 원하는 언어로 번역하는 함수입니다.
    기본 언어는 한국어(ko)로 설정되어 있습니다.
    """

    # 번역기를 실행해서 translator라는 번역기 객체를 만듭니다.
    translator = Translator()

    # 번역기에게 번역 요청을 보냅니다.
    # 번역할 문장은 joke_text이고, 번역될 언어는 lang입니다.
    # *** 번역 결과 텍스트 속성은 .text 입니다. (수정됨) ***
    translated_text = translator.translate(joke_text, dest=lang).text

    # 번역된 문장을 결과로 돌려줍니다.
    return translated_text


# 3. 실제 실행 부분
# 지금까지 만든 함수들을 이용해 유머를 가져오고, 번역하는 과정을 수행

# [기본 영어 유머 요청 및 출력]
# 카테고리를 지정하지 않았으므로, 모든 종류의 유머 중에서 하나를 가져옵니다.
joke_in_english = fetch_joke_from_api()

# 가져온 유머를 영어로 출력합니다.
print("영어 유머:", joke_in_english)

# 가져온 영어 유머를 한국어로 변역합니다.
joke_in_korean = translate_joke_to_korean(joke_in_english)

# 번역된 한국어 유머를 출력합니다.
print("유머 (한국어 번역):", joke_in_korean)
````

````python
# [프로그래밍 카테고리 유머 요청 및 출력]
# 이번에는 category에 "Programming"을 전달해서 프로그래밍 관련 유머를 요청합니다.
joke_in_english = fetch_joke_from_api(category="Programming")

# 가져온 유머를 영어로 출력합니다.
print("영어 유머:", joke_in_english)

# 가져온 프로그래밍 유머를 한국어로 번역합니다.
joke_in_korean = translate_joke_to_korean(joke_in_english)

# 번역된 한국어 유머를 출력합니다.
print("유머 (한국어 번역):", joke_in_korean)


# 같은 유머를 일본어로도 번역해보겠습니다.
# 번역 함수 호출 시 lang="ja"를 전달하면 번역 언어가 일본어로 변경됩니다.
joke_in_korean = translate_joke_to_korean(joke_in_english, lang="ja")

# 번역된 일본어 유머를 출력합니다.
print("유머: (일본어 번역):", joke_in_korean)
````

### 가변 매개변수

- 함수가 **여러 개의 인자를 한꺼번에 받을 수 있게** 해주는 기능
- 개수가 정해지지 않은 **다양한 매개변수를 유연하게 처리**할 때 사용, (*) 기호로 표현

```python
def 함수(값1, 값2):
    print(값1, 값2)
```
**-> 이 함수는 딱 2개만 받을 수 있음.**

```python
def 함수(*값들):
    print(값들)
```
**-> 하지만 가변 매개변수를 쓰면, 2개, 5개, 100개도 마음대로 받을 수 있음.**

#### 가변 매개변수의 구조

```python
def 함수이름(*매개변수이름):
    함수가 해야할 일
```
- 가변 매개변수를 사용하면 개수가 정해지지 않은 여러 개의 값을 받을 수 있다.

**택배 상자 비유**
- 고정 매개변수 = "정해진 칸만 있는 상자" (2개만 담기 가능)
- 가변 매개변수 = "몇 개든 담을 수 있는 무제한 상자"

````python
# 1) 고정 매개변수 함수 예제
# 이 함수는 딱 2개만 값을 받을 수 있음. 
# 두 개이 물건을 받아서 출력하는 함수

def 고정상자(물건1, 물건2):
    # 이 함수는 물건 2개를 받아서 출력하는 함수입니다.
    print(f"첫 번째 물건: {물건1}")
    print(f"두 번째 물건: {물건2}")

# 함수 실행 예시 (물건을 2개 넣어야 합니다)
고정상자("사과", "바나나")
````

````python
# 2) 가변 매개변수 함수 예제
# 이 함수는 몇 개를 받을지 정해두지 않았음.
# 2개를 받을 수도 있고, 5개를 받을 수도 있고, 100개를 받을 수도 있음.

def 가변상자(*물건들):
    """
    이 함수는 '상자에 물건들 담는 함수'입니다.
    담을 물건이 몇 개일지는 미리 정하지 않습니다.
    원하는 만큼 물건을 넣을 수 있습니다.
    물건들은 전부 '물건들'이라는 이름표가 붙은 보따리에 담깁니다.
    """
    print("상자에 담긴 물건 목록:")

    # 받은 물건들을 하나씩 꺼내서 화면에 보여줍니다.
    for 물건 in 물건들:
        print(f"- {물건}")

# 함수 실행 예시 (원하는 만큼 물건을 넣고 실행해봅니다)
가변상자("사과", "바나나", "오렌지")             # 물건 3개
가변상자("연필", "공책")                        # 물건 2개
가변상자("노트북", "마우스", "키보드", "스피커")  # 물건 4개
````

#### 소개글 만들기 예제

- "이름, 나이, 취미를 받아서 자기소개를 출력하는 함수"
- 취미는 **몇 개가 될지 정해두지 않고**, 필요한 만큼 자유롭게 받을 수 있도록 만듦.
- 이름과 나이는 꼭 적어야 하는 필수 항목이고, 취미는 원하는 만큼 몇 개든 적을 수 있는 '추가 항목'

````python
def 소개글만들기(이름, 나이, *취미들):
    """
    이 함수는 '자기소개'를 만드는 함수입니다.
    - 이름과 나이는 꼭 받아야 합니다.
    - 취미는 몇 개를 받을지 미리 정하지 않고, 원하는 만큼 받을 수 있습니다.
    - 취미들은 '취미들'이라는 보따리에 저장됩니다.
    """

    # 이름과 나이를 출력합니다.
    print(f"안녕하세요, 저는 {이름}입니다.")
    print(f"나이는 {나이}살입니다.")

    # 취미가 있는 경우만 출력합니다.
    if 취미들:
        print("제 취미는 다음과 같습니다:")

        # 취미들을 하나씩 꺼내서 출력합니다.
        for 취미 in 취미들:
            print(f"- {취미}")
    else:
        print("취미는 아직 없습니다.")

    # 줄바꿈으로 마무리
    print()

# 함수 실행 예시 (취미 여러 개)
소개글만들기("철수", 25, "축구", "독서", "게임")

# 함수 실행 예시 (취미 2개)
소개글만들기("영희", 30, "요리", "등산")

# 함수 실행 예시 (취미 없음)
소개글만들기("민수", 22)
````

## 리스트

- **여러 개의 데이터를 한 번에 저장**할 수 있는 구조
- [](대괄호)를 사용하여 생성하며, 문자열, 숫자, 불리언 등 다양한 데이터 타입을 하나의 리스트에 함께 저장할 수 있음.
- 리스트를 사용하면 **한 줄**로 여러 개의 데이터를 저장할 수 있음.

```less
리스트: ["사과", "바나나", "오렌지"]

 ┌──────┬────────┬────────┐
 │ 사과  │  바나나 │  오렌지 │
 └──────┴────────┴────────┘
   [0]     [1]      [2]
```

- 리스트에서 값은 **왼쪽부터 0번, 1번, 2번, ...** 이런 식으로 **번호(인덱스)**가 붙는다.
- 이 번호를 사용해서 값을 가져오거나 변경할 수 있다.

### 리스트의 기본 사용법

1. 리스트 만들기

2. 리스트에서 값 가져오기
- **인덱스(번호)**를 사용해서 원하는 값 가져올 수 있다.
- 리스트 인덱스는 **0부터 시작** (0이 첫 번째 값, 1이 두 번째 값)

3. 리스트 값 변경하기
- 리스트 안의 값은 자유롭게 **바꿀 수 있다 !**
- 리스트 값은 인덱스로 접근하여 변경 가능

4. 리스트에 값 추가하기
- 리스트에 값 추가할 때는 **append()** 사용
- append()는 항상 **리스트의 맨 끝**에 값 추가

5. 리스트에서 값 삭제하기
- 필요 없는 값을 리스트에서 **제거**할 수도 있다.
- **remove()**는 리스트에서 특정 값을 찾아 제거
- 같은 값이 여러 개면 첫 번째 값만 제거
- 없는 값을 삭제하려 하면 에러 발생 (ValueError)

````python
# 최종 코드 & 실행 결과

# 1. 리스트 만들기
fruits = ["사과", "바나나", "오렌지"]  # 과일 3개 저장
print("처음 리스트:", fruits)
# 실행 결과: 처음 리스트: ['사과', '바나나', '오렌지']

# 2. 리스트에서 값 가져오기
print("첫 번째 과일:", fruits[0])  # 인덱스 0 -> '사과' 출력
# 실행 결과: 첫 번째 과일: 사과

# 3. 리스트 값 변경하기
fruits[2] = "포도"  # 세 번째 값 '오렌지'를 '포도'로 변경
print("변경 후 리스트:", fruits)
# 실행 결과: 변경 후 리스트: ['사과', '바나나', '포도']

# 4. 리스트에 새로운 값 추가하기
fruits.append("수박")  # 리스트 끝에 '수박' 추가
print("추가 후 리스트:", fruits)
# 실행 결과: 추가 후 리스트: ['사과', '바나나', '포도', '수박']

# 5. 리스트에서 값 제거하기
fruits.remove("바나나")  # '바나나' 삭제 (첫 번째로 찾은 '바나나'만 삭제)
print("삭제 후 리스트:", fruits)
# 실행 결과: 삭제 후 리스트: ['사과', '포도', '수박']
````

### 장바구니 예제

- 파이썬의 **리스트**를 사용하여 **장바구니 프로그램** 만들기
- 이 프로그램은 **3가지 기능**을 제공함.
    - 1) 장바구니에 물건 추가하기
    - 2) 장바구니에서 물건 삭제하기
    - 3) 장바구니에 담긴 물건 확인하기

#### len() 함수
- **리스트의 길이(저장된 데이터 개수)**를 알려주는 함수

````python
# 장바구니 예제 구현

# 🛒 장바구니 리스트를 생성합니다. (처음에는 아무 물건도 들어있지 않습니다.)
cart = []

# 프로그램을 계속 실행하는 부분입니다.
while True:
    # 사용자에게 보여줄 메뉴를 출력합니다.
    print("\n===== 장바구니 프로그램 =====")
    print("1. 물건 추가")  # 장바구니에 물건을 넣는 기능입니다.
    print("2. 물건 삭제")  # 장바구니에서 물건을 빼는 기능입니다.
    print("3. 장바구니 보기")  # 장바구니에 어떤 물건이 있는지 확인하는 기능입니다.
    print("4. 프로그램 종료")  # 프로그램을 종료하는 기능입니다.

    # 사용자에게 어떤 기능을 사용할 지 물어봅니다.
    choice = input("원하는 기능의 번호를 입력하세요: ")

    # 1. 물건을 장바구니에 추가하는 기능입니다.
    if choice == "1":
        item = input("추가할 물건의 이름을 입력하세요: ")  # 어떤 물건을 추가할 지 물어봅니다.
        cart.append(item)  # 사용자가 입력한 물건을 장바구니(cart)에 넣습니다.
        print(f"✅ '{item}이(가) 장바구니에 추가되었습니다.")
        # 현재 장바구니에 들어있는 물건의 개수를 보여줍니다.
        print(f"🛒 현재 장바구니에는 {len(cart)}개의 물건이 있습니다.")

    
    # 2. 물건을 장바구니에서 삭제하는 기능입니다.
    elif choice == "2": 
        # 장바구니가 비어있는 경우, 삭제할 물건이 없다는 메시지를 보여줍니다.
        if len(cart) == 0:
            print("❌ 삭제할 물건이 없습니다!")
        else:
            # 삭제할 물건 이름을 물어봅니다.
            item = input("삭제할 물건의 이름을 입력하세요: ")
            # 사용자가 입력한 물건이 실제 장바구니에 있는지 확인합니다.
            if item in cart:
                cart.remove(item)  # 해당 물건을 장바구니에서 삭제합니다.
                print(f"🗑️ '{item}이(가) 장바구니에서 삭제되었습니다.")
                # 삭제 후, 현재 장바구니에 남은 물건 개수를 보여줍니다.
                print(f"🛒 현재 장바구니에는 {len(cart)}개의 물건이 있습니다.")
            else:
                # 장바구니에 없는 물건을 삭제하려고 했을 때, 안내 메시지를 보여줍니다.
                print("❌ 장바구니에 없는 물건입니다!")
    
    
    # 3. 장바구니에 있는 물건들을 확인하는 기능입니다.
    elif choice == "3":
        # 장바구니가 비어있을 때와 물건이 있을 때 각각 다른 메시지를 보여줍니다.
        if len(cart) == 0:
            print("🛒 장바구니가 비어 있습니다.")
        else:
            # 장바구니에 있는 모든 물건 목록을 보여줍니다.
            print("🛍️ 현재 장바구니 내용:", cart)
            # 현재 장바구니에 들어있는 물건 개수도 함께 보여줍니다.
            print(f"🛒 현재 장바구니에는 {len(cart)}개의 물건이 있습니다.")
    

    # 4. 프로그램을 종료하는 기능입니다.
    elif choice == "4":
        print("👋🏻 프로그램을 종료합니다. 감사합니다!")
        break  # 반복문(while)을 끝내고 프로그램을 종료합니다.

    # 1~4번 이외의 잘못된 번호를 입력했을 때 실행되는 부분입니다.
    else:
        print("❌ 잘못된 입력입니다! 1~4번 중에서 선택해주세요.")
````
