### 리스트 컴프리헨션 (List Comprehension)

- **쉽고 빠르게 리스트를 만드는 방법**
- 한 줄의 코드로 여러 줄의 반복문 대체 가능

#### 리스트 컴프리헨션의 기본 구조

```python
[값 for 변수 in 반복문 if 조건]
```
- 값: 리스트에 넣을 값
- 변수: 반복문에서 사용할 변수
- 반복문: 값을 만들 때 사용할 반복문
- if 조건 (선택 사항): 특정 조건을 만족하는 값만 리스트에 추가

**리스트를 만드는 두 가지 방법**
1. 일반적인 리스트 생성 방법
- 손으로 하나씩 리스트에 추가하는 방식
    - [] 리스트를 먼저 만들고,
    - for 반복문을 사용해서 값을 추가
    - 리스트에 값을 넣기 위해 append() 함수를 여러 번 사용해야 함.

````python
numbers = []  # 빈 리스트 만들기

for i in range(1, 6):  # 1부터 5까지 반복
    numbers.append(i)  # 리스트에 숫자 추가

print(numbers)  # 출력 결과: [1, 2, 3, 4, 5]
````

2. 리스트 컴프리헨션을 사용한 리스트 생성
- 한 줄의 코드로 리스트를 만들 수 있다 !

````python
# numbers라는 이름의 리스트 만들기.
# 리스트란, 여러 개의 데이터를 한꺼번에 보관할 수 있는 상자.
# 이 리스트에는 1부터 5까지 숫자를 순서대로 담을 예정.

# 아래 코드는 '리스트 컴프리헨션'이라는 문법을 사용한 것.
# 리스트 컴프리헨션은 '리스트를 한 줄로 간단하게 만드는 방법'
# 여기서 [i for i in range(1, 6)]의 의미를 설명.

# range(1, 6)는 1부터 5까지의 숫자를 순서대로 만들어준다.
# (6은 포함하지 않아서 1, 2, 3, 4, 5만 나온다.)

# for i in range(1, 6)는 이 숫자들을 하나씩 꺼내오면서
# i라는 이름으로 사용하겠다는 뜻.

# i for i in range(1, 6)는 
# 꺼낸 숫자(i)를 그대로 리스트에 담는다는 뜻.
# 결과적으로 numbers라는 리스트에는 [1, 2, 3, 4, 5]가 저장됨.
numbers = [i for i in range(1, 6)]

# 완성된 리스트(numbers)를 화면에 출력.
print(numbers)  # 출력 결과: [1, 2, 3, 4, 5]
````

````python
# 숫자를 2배로 만든 리스트

# 1. 일반적인 리스트 사용
numbers1 = []

for i in range(1, 6):
    numbers1.append(i * 2)  # i에 2를 곱한 값을 리스트에 추가

print(numbers1)


# 2. 리스트 컴프리헨션 사용
numbers2 = [i * 2 for i in range(1, 6)]  # 반복문이 리스트 안에 들어가서 코드가 짧아진다 !

print(numbers2)
````

````python
# 짝수만 리스트에 저장하기

# 1. 일반적인 리스트 사용
even_numbers1 = []

for i in range(1, 11):
    # 짝수는 2로 나누었을 때 나머지가 0인 숫자
    if i % 2 == 0:
        even_numbers1.append(i)

print(even_numbers1)


# 2. 리스트 컴프리헨션 사용
even_numbers2 = [i for i in range(1, 11) if i % 2 == 0]  # if 조건문을 추가하면 특정 값만 리스트에 저장 가능 !

print(even_numbers2)
````

````python
# 양수 음수 판별기

# 1. 일반적인 리스트 사용
scores = [10, -5, 20, -15, 0]

status1 = []

for s in scores:
    # 점수가 0보다 크면 양수로 판단
    if s > 0:
        status1.append("Positive")  # 양수면 "Positive" 추가
     
    elif s < 0:
        status1.append("Negative")
    
    else:
        status1.append("Zero")

print(status1)


# 2. 리스트 컴프리헨션 사용
status2 = [
    "Positive" if s > 0 else "Negative" if s < 0 else "Zero"
    for s in scores
]

print(status2)
````

## 객체

- 파이썬은 객체 지향 프로그래밍 언어로, 모든 것이 객체.
- 객체는 데이터와 기능을 하나로 모은 프로그래밍 단위.
- 객체 사용의 장점은 코드 재사용성, 프로그램 구조화, 실제 세계의 개념을 쉽게 표현할 수 있다는 점.
- 우리 일상에서 볼 수 있는 모든 사물이나 개념을 의미함. (ex. 자동차, 휴대폰, 책, 강아지 등)

### 클래스

- 객체를 만들기 위한 설계도

### 생성자

- 객체를 만들 때 자동으로 호출되는 특별한 함수
- 생성자의 역할은 객체가 생성될 때 필요한 초기 값을 설정하는 것.

### 클래스 선언

- 설계도 자체를 만드는 과정

#### 파이썬 클래스 선언 기본 구조

1) 클래스 이름 정하기

```python
class Car:
```

- "나는 지금부터 클래스를 만들 거야"라고 컴퓨터에게 알려주는 명령어 (class)
- Car는 이 클래스(설계도)의 이름

2) 생성자 만들기 (초기 정보 설정하기)

```python
def __init__(self, color, brand, max_speed):
```

- __init__은 특별한 이름으로, **처음 만들어질 때 실행되는 함수**
- ()안의 정보는 이 설계도를 만들 때 필요한 정보를 외부에서 받아오겠다는 의미.

3) 속성 저장하기 (가진 정보 저장하기)

```python
self.color = color
self.brand = brand
self.max_speed = max_speed
```

- 받아온 정보를 설계도가 기억할 수 있도록 저장해둬야 하는데, 이때 사용하는 것이 속성.

````python
# 자동차(Car) 클래스를 정의합니다.
# 클래스는 '자동차'라는 개념을 프로그램 안에서 표현하는 방법입니다.
# 자동차의 색상, 브랜드, 최고 속도 같은 정보를 저장하고 관리할 수 있습니다.

class Car:  # Car라는 이름의 클래스를 정의합니다.
    # __init__ 메서드는 "생성자"라고 부릅니다.
    # 생성자는 자동차 객체가 만들어질 때 자동으로 실행되는 특수한 함수입니다.
    # 생성자는 자동차의 색상, 브랜드, 최고 속도를 설정하는 역할을 합니다.
    def __init__(self, color, brand, max_speed):
        # self는 '현재 만들어지고 있는 자동차 객체 자신'을 의미합니다.
        # self를 통해 '이 자동차 객체만의 속성'을 설정할 수 있습니다.

        # color, brand, max_speed는 생성자를 호출할 때 외부에서 전달하는 값입니다.
        # 즉, 자동차의 색상, 브랜드, 최고 속도 정보를 외부에서 넘겨주는 값입니다.

        # self.color, self.brand, self.max_speed는 '자동차 객체가 실제로 가지는 속성'입니다.
        # 각각의 자동차 객체마다 이 값들은 다를 수 있습니다.

        # 즉, color는 생성자에서 잠깐 사용하는 '임시 값'이고,
        # self.color는 이 자동차 객체가 '평생 동안 기억할 진짜 속성 값'입니다.
        self.color = color  # 전달받은 색상(color)을 자동차의 'color 속성'에 저장합니다.
        self.brand = brand  # 전달받은 브랜드(brand)를 자동차의 'brand 속성'에 저장합니다.
        self.max_speed = max_speed  # 전달받은 최고 속도(max_speed)를 자동차의 'max_speed 속성'에 저장합니다.

        # 정리하면
        # color -> 자동차를 만들 때 잠깐 쓰는 값 (외부에서 가져온 값)
        # self.color -> 이 자동차가 계속 가지고 있을 진짜 속성 값 (객체의 내부 값)
````


````python
# 자동차 객체 생성
# 'my_car'라는 이름의 자동차 객체를 만듭니다.
# 이 자동차의 색상은 "빨간색", "현대", 200이 바로 color, brand, max_speed로 전달됩니다.
my_car = Car("빨간색", "현대", 200)
````

#### 속성

- 객체가 가지고 있는 **특징(정보)**

````python
# 자동차 속성 출력
# 만들어진 자동차의 정보를 화면에 출력하는 부분입니다.
# 자동차 객체에 저장된 색상, 브랜드, 최고 속도를 확인할 수 있습니다.

print("자동차 색상:", my_car.color)  # 자동차의 'color 속성'을 출력합니다.
print("자동차 브랜드:", my_car.brand)  # 자동차의 'brand 속성'을 출력합니다.
print("자동차 최고 속도:", my_car.max_speed)  # 자동차의 'max_speed 속성'을 출력합니다.
````

```python
객체명.속성명 = 새로운 값
```

#### 메서드

- 객체가 수행할 수 있는 **행동(기능)**

```python
def 메서드이름(self):
    # 여기서 기능을 만듭니다.
```

- def: "나는 지금 기능을 만들 거야"
- 메서드이름: 기능의 이름
- self: 이 메서드가 어떤 설계도에서 실행되는지 알려주는 특별한 키워드

````python
# 자동차를 운전하거나 멈추는 기능도 포함할 수 있습니다.

class Car:
    def __init__(self, color, brand, max_speed):
        self.color = color
        self.brand = brand
        self.max_speed = max_speed

    # 자동차가 운전하는 기능을 담당하는 메서드입니다.
    # 메서드는 자동차가 할 수 있는 동작(행동)을 정의하는 함수
    def drive(self):
        # 자동차가 달리는 기능을 출력하는 코드
        # 자동차 브랜드와 최고 속도를 출력하여 자동차가 어떤 속도로 달리는지 보여줍니다.
        print(f"{self.brand} 자동차가 {self.max_speed}km/h로 달립니다.")

    # 자동차를 멈추는 기능을 담당하는 메서드
    def stop(self):
        # 자동차가 멈추는 기능을 출력하는 코드
        print(f"{self.brand} 자동차가 멈췄습니다.")
````

````python
# 자동차 객체 생성
# 'my_car'라는 이름의 자동차 객체를 만듭니다.
# 이 자동차의 색상은 "빨간색", 브랜드는 "현대", 최고 속도는 200으로 설정합니다.
# 이때 전달되는 "빨간색", "현대", 200이 바로 color, brand, max_speed로 전달됩니다.
my_car = Car("빨간색", "현대", 200)

# 자동차 기능 실행 (메서드 실행)
# 자동차가 달리게 합니다.
my_car.drive()  # 자동차의 drive 메서드를 실행합니다.

# 자동차가 멈추게 합니다.
my_car.stop()  # 자동차의 stop 메서드를 실행합니다.
````

### 리턴 값이 있는 메서드

- 리턴 값이 있는 메서드는 **결과 값을 반환**합니다.

````python
# 예제: 원의 넓이는 계산하는 메서드

# 원(Circle) 클래스를 정의합니다.
# 클래스는 '원'이라는 개념을 코드로 표현하는 방법입니다.
# 원의 반지름을 저장하고, 원의 넓이를 계산하는 기능을 포함합니다.

class Circle:  # Circle이라는 이름의 클래스를 정의합니다.
    # __init__ 메서든느 "생성자"라고 부릅니다.
    # 생성자는 원 객체가 처음 만들어질 때 자동으로 실행되는 특별한 함수입니다.
    # 이 생성자에서는 원의 반지름을 설정하는 역할을 합니다.
    def __init__(self, radius):
        # self는 '현재 만들어지고 있는 원 객체 자신'을 의미합니다.
        # self를 사용하면 각 원 객체가 서로 다른 반지름 값을 가질 수 있습니다.

        # radius는 원을 만들 때 외부에서 전달받는 값입니다.
        # 즉, 원의 반지름 정보를 외부에서 넘겨주는 값입니다.

        # self.radius는 '이 원 객체가 실제로 가지는 반지름 속성'입니다.
        # 각각의 원 객체마다 반지름이 서로 다를 수 있습니다.
        self.radius = radius  # 원의 반지름 정보를 저장하는 속성입니다.

    # 원의 넓이를 계산하는 메서드입니다.
    # 메서드는 원이 할 수 있는 동작(기능)을 정의하는 함수입니다.
    def calculate_area(self):
        # 원의 넓이를 계산하는 공식은 π * 반지름 * 반지름입니다.
        # 여기서는 π 값을 3.14로 사용합니다.
        return 3.14 * self.radius * self.radius
    

# 원 객체 생성
# 'circle1'이라는 이름의 원 객체를 만듭니다.
# 이 원의 반지름은 5로 설정합니다.
circle1 = Circle(5)

# 원의 넓이 계산
# circle1 객체의 calculate_area() 메서드를 실행하여 넓이를 계산합니다.
area = circle1.calculate_area()

# 원의 반지름과 넓이를 출력
# 계산된 넓이를 포함하여 원의 반지름과 함께 출력합니다.
print(f"반지름이 {circle1.radius}인 원의 넓이: {area}")
````

#### 쇼핑몰 상품 관리 예제

- 객체 지향 프로그래밍을 활용해 실제 쇼핑몰의 상품 관리 시스템 구현해보기

**객체를 활용하는 이유**
- 쇼핑몰에는 여러 개의 **상품(객체)**이 있음.
- 각 상품은 **이름, 가격, 재고(수량)**을 가지고 있고, **가격을 할인하거나 재고를 감소시키는 행동(메서드)**를 할 수 있음.
- 즉, **"상품"이라는 객체를 만들고, 상품의 가격을 변경하거나 재고를 줄이는 기능을 추가**할 수 있음.

````python
# 상품 객체를 만들고 정보를 출력하는 코드

# 상품(Product) 클래스를 정의합니다.
# 클래스는 '상품'이라는 개념을 코드로 표현하는 방법
# 상품의 이름, 가격, 재고 수량을 저장하고 관리 가능

class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock
    
    # 상품 정보를 출력하는 메서드
    def show_info(self):
        print(f"상품명: {self.name}, 가격: {self.price}원, 재고: {self.stock}개")

    
    # 상품 가격 할인 기능 추가
    def apply_discount(self, discount):
        # 현재 가격에서 할인 금액을 빼서 새로운 가격을 계산합니다.
        self.price -= discount

        print(f"{self.name}의 가격이 {discount}원 할인되어 {self.price}원이 되었습니다.")

    # 상품을 구매 시 재고 감소 기능 추가
    def buy_product(self, quantity):
        # 구매하려는 수량이 재고보다 적거나 같으면 구매가 가능합니다.
        if quantity <= self.stock:
            self.stock -= quantity
            print(f"{quantity}개 구매 완료 ! 남은 재고: {self.stock}개")
        else:
            print('재고가 부족합니다.')


# 상품 객체 생성
product1 = Product("노트북", 1200000, 10)

# 기존 상품 정보 출력
print("===== 기존 상품 정보 =====")
product1.show_info()


# 할인 적용
print("===== 할인 적용 =====")
product1.apply_discount(100000)

# 할인 후 상품 정보 출력
print("===== 할인 후 상품 정보 =====")
product1.show_info()

# 상품 구매
print("===== 상품 구매 (2개) =====")
product1.buy_product(2)

# 다시 상품 정보 출력
print("===== 구매 후 상품 정보 =====")
product1.show_info()

# 재고 부족 테스트
print("===== 재고 부족 테스트 (10개 구매 시도) =====")
product1.buy_product(10)
````

## 딕셔너리

- 키와 값이 한 쌍으로 저장되는 데이터 구조, **이름표를 붙여 정보를 저장하는 방식**
- 실제 사전처럼 키를 통해 원하는 정보를 빠르게 찾을 수 있음.
- {}(중괄호)로 만들고, 순서는 없으며 같은 키를 두 번 쓸 수 없음.

### 딕셔너리 기본 문법

```python
딕셔너리이름 = {
    "이름표1": 값1,
    "이름표2": 값2,
    "이름표3": 값3
}
```

- **이름표(key)**와 **값(value)**을 **콜론(:)으로 연결함.

````python
# 딕셔너리 생성하기

# 이 코드는 사람 한 명의 정보를 저장하는 예제
# '이름', '나이', '직업'이라는 3가지 정보를 저장하고 있음.

# 사람 정보를 저장하는 '딕셔너리'를 만듦.
# 딕셔너리는 여러 정보를 하나로 묶어서 보관하는 자료형
# 키(key)와 값(value) 형태로 데이터를 저장
# 예를 들어, '이름'이라는 키에는 '홍길동'이라는 값이 저장됨.

person = {
    "이름": "홍길동",  # 이름이라는 키에 '홍길동'이라는 값을 저장
    "나이": 25,  # 나이라는 키에 숫자 25 저장
    "직업": "개발자"  # 직업이라는 키에 '개발자'라는 값을 저장
}

# 저장한 사람 정보를 화면에 출력합니다.
print(person)

# 딕셔너리 전체가 출력되어서, 저장된 내용을 한 눈에 볼 수 있음.
````

````python
# 딕셔너리에서 값 가져오기
# 딕셔너리에서 특정 값만 가져오려면 이름표(key)를 사용함.

# 이 코드는 'person'이라는 딕셔너리에 저장된 정보를 꺼내서 출력하는 예제
# '딕셔너리'란 여러 정보를 한 곳에 모아두고, 이름표(키)를 이용해 쉽게 찾을 수 있는 저장 방식
# 예를 들어 '이름'이라는 키로 '홍길동'이라는 값을 꺼낼 수 있음.

# '이름'이라는 키에 해당하는 값을 출력함.
# 즉, 홍길동이라는 이름이 화면에 표시됨.
print(person["이름"])  # 홍길동이 출력됨.

# '나이'라는 키에 해당하는 값을 출력함.
# 즉, 숫자 25가 화면에 표시됨.
print(person['나이'])  # 25가 출력됨.
````

````python
# 없는 키를 입력하면 오류 발생

# 이 코드는 'person'이라는 딕셔너리에서 '주소'라는 정보를 꺼내려고 함.
# 그런데 '주소'라는 정보는 person 안에 없음.
# 그래서 프로그램이 오류를 내면서 멈춤.
# 이런 오류를 'KeyError(키 에러)'라고 부름.
# 존재하지 않는 정보를 찾을 때 발생하는 오류

# print(person['주소'])  # 오류 발생 (KeyError)


# 해결 방법

# 이번에는 'person'에서 '주소'라는 정보를 꺼내되,
# 만약 '주소'가 없다면 '정보 없음'이라는 기본 문구를 대신 보여주게 함.
# 이렇게 하면 오류가 발생하지 않고, 프로그램이 멈추지 않음.
# 'get()'은 딕셔너리에서 정보를 안전하게 찾을 때 사용하는 기능
# 두 번째로 적은 '정보 없음'이 기본값으로 표시되는 내용

print(person.get("주소", "정보 없음"))
````

````python
# 딕셔너리에 새로운 데이터 추가

# 이 코드는 'person'이라는 딕셔너리에 새로운 정보를 추가하는 예제
# 지금까지 '이름', '나이', '직업'만 있었는데, 여기에 '주소'라는 정보를 새로 넣으려고 함.

# '주소'라는 키에 '서울'이라는 값을 저장함.
# 즉, '주소'라는 이름표를 새로 만들고, 그 안에 '서울'이라는 내용을 적는 것과 같음.
person['주소'] = '서울'

# 이제 딕셔너리에 '주소' 정보가 추가되었으니, 전체 내용을 출력해보겠습니다.
print(person)
````

````python
# 딕셔너리에서 데이터 삭제

# 이 코드는 'person'이라는 딕셔너리에 저장된 정보 중에서
# '직업'이라는 정보를 삭제하는 예제
# 'del'이라는 명령어는 딕셔너리에서 특정 정보를 완전히 지울 때 사용함.

# '직업'이라는 키와 그에 해당하는 값을 딕셔너리에서 삭제함.
# 삭제된 정보는 다시 확인하거나 사용할 수 없음.
del person['직업']

# 남아있는 정보를 확인하기 위해 딕셔너리를 출력해봅니다.
print(person)
````

### 딕셔너리 반복문

- 반복문을 이용하면 딕셔너리에 저장된 모든 데이터를 **자동으로 하나씩 꺼내서 처리할 수 있음.**

#### 딕셔너리에서 반복문 구조

```python
for 변수 in 딕셔너리:
    실행할 코드
```

- 딕셔너리는 기본적으로 키(Key)만 반복함.
- 값(value)만 가져오려면 .values()를 사용함.
- 키와 값을 함께 가져오려면 .items()를 사용함.
    - 키와 값을 **튜플(쌍)**로 가져옴.

````python
# 딕셔너리의 키(Key)만 가져오는 반복문

scores = {
    "철수": 85,
    "영희": 92,
    "민수": 78
}

# 반복문을 사용하여 딕셔너리에 있는 모든 학생 이름(키)만 순서대로 출력하는 부분
# 'for'문은 같은 동작을 여러 번 반복할 때 사용하는 명령어
# scores에 저장된 학생 이름을 하나씩 꺼내서 name이라는 변수에 담고, 그 이름을 출력하는 과정
for name in scores:
    print(name)
````

````python
# 딕셔너리에서 값(Value)만 가져오는 반복문

scores = {
    "철수": 85,
    "영희": 92,
    "민수": 78
}

# 'scores.values()'는 딕셔너리에서 점수들(값)만 가져오는 기능
for score in scores.values():
    print(score)
````

````python
# 딕셔너리에서 키(Key)와 값(Value) 모두 가져오는 반복문

scores = {
    "철수": 85,
    "영희": 92,
    "민수": 78
}

# 'scores.items()'는 딕셔너리에 저장된 이름표(키)와 값(점수)을 한 쌍씩 가져오는 기능
for name, score in scores.items():
    print(f"{name}의 점수는 {score}점입니다.")
````

````python
# 음식 주문 목록 관리 예제

# 이 코드는 손님들의 음식 주문 내용을 저장하고, 출력하는 프로그램.
# 'orders'라는 딕셔너리에 손님 이름과 주문한 음식 목록을 보관하고 있음.
# 딕셔너리 안에 또 다른 딕셔너리가 들어있는 '딕셔너리 속 딕셔너리' 구조

orders = {
    "김철수": {"김치찌개": 1, "불고기": 2},
    "이영희": {"된장찌개": 1, "비빔밥": 1}
}

# 모든 손님의 주문 내용을 출력
for customer, menu in orders.items():
    print(f"{customer}님의 주문 내역:")

    # 이번에는 해당 손님이 주문한 음식과 개수를 출력
    for food, quantity in menu.items():
        print(f"- {food}: {quantity}개")

    # 줄 바꿈
    print()
````

#### 최근 3개월 미구매 고객 찾기

- **가게에서 오랫동안 안 온 손님을 찾아내는 프로그램**
    - 가게가 고객들의 "마지막 구매 날짜"를 기록해 두었음.
    - 오늘 날짜를 기준으로, 90일 동안 아무런 구매 기록이 없는 고객을 "이탈 고객"이라고 부름.
    - 이 프로그램은 "이탈 고객이 누구인지"를 찾아서 화면에 보여줌.

**순서)**
1. 고객들의 이름과 마지막 구매 날짜를 기록해둔 자료를 준비
2. 오늘 날짜 또는 기준 날짜를 정함.
3. 모든 고객의 구매 날짜와 기준 날짜의 차이를 계산
4. 그 차이가 90일보다 크면 "이탈 고객"으로 판단
5. 이탈 고객 명단을 화면에 보여줌.

````python
# 날짜와 시간을 다루는 datetime 라이브러리 불러오기
# 날짜끼리의 차이를 구하거나, 날짜 문자열을 실제 날짜로 변환할 때 필요.
from datetime import datetime


# 고객들의 마지막 구매 날짜 정보를 담고 있는 리스트
# 리스트 안에는 각 고객의 정보를 저장한 딕셔너리
orders = [
    {"고객명": "철수", "구매일": "2024-01-01"},
    {"고객명": "영희", "구매일": "2024-10-10"}
]

# 이탈 고객을 판단한 기준 날짜
# '2024-03-01'이라는 문자열을 실제 날짜 데이터(datetime 객체)로 변환
# strptime() 함수는 문자열(String)을 시간(Time)으로 바꿔주는 함수
# strptime()에 있는 "%Y-^m-%d"는 날짜 형식을 의미
# "&Y"는 연도(4자리), "%m"은 월(2자리), "%d"는 일(2자리)를 의미
cutoff = datetime.strptime("2024-05-01", "%Y-%m-%d")

# 90일 넘게 구매 이력이 없는 고객 이름을 담아둘 빈 리스트
inactive_customers = []

# 고객들의 주문 기록을 하나씩 확인
for order in orders:
    name = order['고객명']

    # 주문 정보에 있는 '구매일'은 문자열 -> 실제 날짜(datetime 객체)로 변환
    date = datetime.strptime(order['구매일'], "%Y-%m-%d")


    # 기준일(cutoff)에서 구매일(date)을 빼면 날짜 차이가 나옴.
    # (cutoff - date)는 '시간 차이'를 나타내는 timedelta라는 값이 됨.
    # timedelta.days는 그 시간 차이를 '며칠' 단위로 바꿔서 알려줌.
    # 즉, (cutoff - date).days는 '며칠 동안 구매를 안 했는지'를 의미
    print((cutoff - date).days)

    if (cutoff - date).days > 90:
        inactive_customers.append(name)

    
# 최종 결과 출력
print(f"이탈 고객: {inactive_customers}")
````
