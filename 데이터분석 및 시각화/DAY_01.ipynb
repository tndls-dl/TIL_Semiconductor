## 데이터 시각화

### 시각화

데이터를 숫자나 글자로 보는 대신, 그래프나 그림으로 만들어 쉽게 이해할 수 있도록 보여주는 것.

- 시각화가 필요한 이유 : **한 눈에 쉽게 문제를 발견하기 위해서**

#### 대표적인 종류

1. 막대 그래프(Bar Chart)
2. 꺾은선 그래프(Line Chart)
3. 원 그래프(Pie Chart)
4. 히스토그램(Histogram)
5. 산점도(Scatter Plot)
