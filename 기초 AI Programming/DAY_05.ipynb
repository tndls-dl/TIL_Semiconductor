### 피봇 테이블

- 원래 복잡한 표 데이터를 쉽게 정리하고 요약해서 볼 수 있도록 도와주는 기능

#### 피봇 테이블 만드는 법
```python
# 피봇 테이블을 만드는 형식 (기본)
데이터프레임.pivot_table(index='행이 될 열 이름', values='계산할 열 이름', aggfunc='계산방법')
```

- **index**(인덱스): **어떤 기준**으로 데이터를 볼 지 정하는 곳
- **values**(밸류): 어떤 **숫자 데이터를 계산**할 지 정함.
- **aggfunc**(계산방법): 데이터 값을 어떻게 요약할 지 정하는 부분

````python
# 예제

import pandas as pd

data = {
    '손님': ['철수', '영희', '민수', '철수', '영희', '민수'],
    '메뉴': ['피자', '피자', '피자', '햄버거', '햄버거', '햄버거'],
    '개수': [2, 1, 3, 2, 2, 1]
}

df = pd.DataFrame(data)
print(df)
print()


# 메뉴별로 총 개수 계산하기

# 메뉴별로 판매된 개수를 합산해서 새로운 표로 만들기
pivot_menu = df.pivot_table(
    index='메뉴',
    values='개수',
    aggfunc='sum'
)

print(pivot_menu)
print()


# 손님별로 총 주문 수량 계산하기

# 손님별 총 주문한 개수 합산
pivot_customer = df.pivot_table(
    index='손님',
    values='개수',
    aggfunc='sum'
)

print(pivot_customer)
print()


# 메뉴별로 평균 계산하기

# 메뉴가 주문될 때마다 평균적으로 몇 개씩 주문하는지 피봇 테이블 만들기
pivot_mean = df.pivot_table(
    index='메뉴',
    values='개수',
    aggfunc='mean'
)

print(pivot_mean)
````

### 크로스탭(crosstab)

- 두 가지 항목 사이의 관계를 쉽게 정리해서 보여주는 표를 만드는 기능

#### crosstab 형식

```python
pd.crosstab(표의 행으로 나올 항목, 표의 열로 나올 항목)
```
- 첫 번째 항목: 표의 가로줄(행)에 나올 항목
- 두 번째 항목: 표의 세로줄(열)에 나올 항목

````python
# 예제

import pandas as pd

data = {
    '성별': ['남자', '여자', '남자', '여자', '남자', '여자', '남자'],
    '메뉴': ['피자', '피자', '햄버거', '햄버거', '피자', '햄버거', '햄버거']
}

df = pd.DataFrame(data)
print(df)
print()


# 성별과 메뉴 관계
# 성별과 메뉴 간의 관계를 정리한 크로스탭 만들기
table = pd.crosstab(df['성별'], df['메뉴'])

print(table)
print()


# 메뉴와 성별 관계 순서 바꾸기
# 메뉴를 가로줄에, 성별을 세로줄에 놓고 다시 크로스탭 만들기
table2 = pd.crosstab(df['메뉴'], df['성별'])

print(table2)
print()


# 합계 표시하기
# 합계를 추가한 크로스탭 만들기 (margins=True는 합계를 표시해 준다.)
table3 = pd.crosstab(df['성별'], df['메뉴'], margins=True)

print(table3)
print()


# 성별에 따른 메뉴 선호도 퍼센트로 보기
# 크로스탭에서 각 메뉴 선택의 비율(퍼센트)을 계산하기
table_percent = pd.crosstab(df['성별'], df['메뉴'], normalize=True) * 100
# normalize=True는 비율로 바꿔준다.

table_percent = table_percent.round(2)

print(table_percent)
print()


# 합계와 퍼센트를 모두 함께 표시하기
# 성별-메뉴 크로스탭에 합계(All)를 표시하고 퍼센트로 나타내기
table_percent_all = pd.crosstab(df['성별'], df['메뉴'], margins=True, normalize='all') * 100
table_percent_all = table_percent_all.round(2)

print(table_percent_all)
````

### 결함이 가장 많은 웨이퍼 10개의 원인 찾기

#### 분석 개요

1. 결함이 없는 데이터 (정상 웨이퍼):
정상적으로 제조된 웨이퍼 700개의 데이터를 분석했습니다. 웨이퍼가 어떤 조건에서 생산됐는지 특징을 살펴보았습니다.

2. 결함이 있는 데이터 (불량 웨이퍼):
결함이 발생한 웨이퍼 300개의 데이터를 분석했습니다. 어떤 조건에서 결함(균열, 핀홀, 입자)이 생겼는지 특징을 파악했습니다.

3. 결함이 가장 많은 데이터 10개 (심각한 불량 웨이퍼):
특별히 결함이 심각한 상위 10개의 웨이퍼를 집중적으로 분석했습니다. 왜 다른 웨이퍼들보다 결함이 많았는지 구체적인 원인은 찾아 봤습니다.

#### 분석 이유
- 불량의 원인 파악
- 품질 관리 개선
- 생산 효율성 증대

#### 기대 효과
- 불량률 감소
- 제조 비용 절감
- 품질 향상

````python
import pandas as pd

oxide_df = pd.read_csv("02_oxide.csv")

# defect_count 값이 0인 데이터만 필터링 (즉, 결함이 없는 웨이퍼만 선택)
no_defect_df = oxide_df[oxide_df['defect_count'] == 0]

# 결함이 없는 행의 통계 분석
no_defect_df.describe()

# defect_count 값이 0보다 큰 데이터만 필터링(즉, 결함이 있는 웨이퍼만 선택)
defect_df = oxide_df[oxide_df['defect_count'] > 0]

# 결함이 있는 행의 통계 분석
defect_df.describe()

# defect_count 값이 많은 순서대로 정렬
defect_df = defect_df.sort_values(by='defect_count', ascending=False)

# 상위 10개 데이터 high_defect_df 대입
high_defect_df = defect_df.head(10)

# 결함이 높은 10개 행의 통계 분석
high_defect_df.describe()
````

#### 결론 (결함 원인 분석)
- 너무 높은 산화 온도와 산화 온도의 급격한 변화가 가장 큰 원인으로 지목됨.

#### 결론 및 효과
- 정상적인 웨이퍼는 약 **998도 근처의 온도에서 안정적으로 생산**되고 있는 반면, 온도가 조금 더 높아지거나 급격하게 변화할 경우 결함 발생 가능성이 높아진다는 사실을 확인할 수 있습니다.

### 시간대별 결함 분석
- 웨이퍼 제조 데이터를 시간대별로 구분하여 오전(6시~12시), 오후(12시~18시), 밤(18시~06시)으로 나누었습니다.
- 각 시간대별로 웨이퍼에 발생한 결함 유형(Crack, Particle, Pinhole)의 개수를 분석하여 어떤 시간대에 어떤 결함이 많이 발생하는지 확인했습니다.

#### 분석 이유
- 시간대별 품질 관리
- 공정 관리 효율성
- 작업자 및 장비 관리 개선

````python
import pandas as pd

oxide_df = pd.read_csv("02_oxide.csv")

# process_date(날짜) + process_time(시간)을 합쳐서 process_datetime컬럼 생성
# pd.to_datetime() -> 데이터를 날짜-시간(datetime) 형식으로 변환
oxide_df['process_datetime'] = pd.to_datetime(oxide_df['process_date'] + ' ' + oxide_df['process_time'])

# 하루를 오전(Morning), 오후(Afternoon), 밤(Night)으로 나누는 함수
def get_time_of_day(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    else:
        return 'Night'
    
# 각 데이터에 시간대 정보 추가 (오전, 오후, 밤)
# dt.hour -> process_datetime에서 시간(hour)만 추출
# .apply(get_time_of_day) -> 추출한 시간을 get_time_of_day() 함수로 변환해 오전, 오후, 밤을 분류
oxide_df['time_of_day'] = oxide_df['process_datetime'].dt.hour.apply(get_time_of_day)

# 결함이 있는 데이터만 선택 (결함 유형이 'none'인 데이터는 제외)
defect_df = oxide_df[oxide_df['defect_type'] != 'none']

# 시간대별로 결함 유형별 결함 수를 피벗 테이블을 이용해 정리
time_defect_analysis = pd.pivot_table(
    defect_df,
    index='time_of_day',
    columns='defect_type',
    values='defect_count',
    aggfunc='sum',
    fill_value=0  # 데이터가 없을 때 0으로 채움
)

time_defect_analysis
````

#### 분석 결과에 대한 해석
- **밤(Night)** 시간대에 모든 결함 유형(Crack, Particle, Pinhole)이 가장 많이 발생함.
- 특히 **Crack(균열)** 유형 결함은 밤에 월등히 많음.
    - 현장에서 가장 빈번히 지적되는 Crack의 원인은 급격한 온도 변화로 웨이퍼 내부에서 불균일한 팽창 및 수축이 발생하여 균열이 생긴 것으로 알려져 있습니다.
    - 이는 야간의 온도 변화 또는 장비 상태가 가장 큰 원인으로 작용했을 가능성이 있습니다.
- 오전과 오후의 결함 수는 밤보다 상대적으로 낮으며 비교적 균일한 수준을 유지하고 있습니다.

---
#### 결함 발생 원인 및 개선 방안
- **밤(Night)**: 설비의 누적된 피로, 유지보수 미흡, 온습도 변화 등 관리가 취약해질 수 있는 시간대로, 이에 대한 설비 유지보수 및 관리를 강화하는 것이 중요.
- **오후(Afternoon)**: 상대적으로 Pinhole 결함이 많아 압력이나 균일성 관리에 좀 더 신경 써야 함.
- **오전(Morning)**: 결함이 가장 적게 발생했지만, 지속적인 모니터링을 유지하여 공정 안정성을 유지하는 것이 좋음.

##### 개선 방안 제안
- 야간(밤) 공정 중 설비 유지보수 및 점검 빈도를 높이고, 작업자의 피로 관리 및 교대 근무를 최적화할 필요가 있음.
- 균열(Crack)이 많으므로 야간 시간대에는 산화 온도 변화를 더 세밀히 관리해야 함.

### Unif(균일도)

**"두께의 일정함"** 을 수치로 나타낸 것

#### Unif(균일도)의 공식

$\text{Unif}(\%) = \frac{\text{최대값} - \text{최소값}}{2 \times \text{평균값}} \times 100$ 

- 가장 두꺼운 부분과 가장 얇은 부분의 차이를 평균 두께로 나누어 얼마나 일정한 두께로 유지되는지 숫자로 나타낸 것.


**최대-최소값 균일성** (Range-based Uniformity) :
- 전체 측정값에서 최대값과 최소값의 차이를 평균값의 2배로 나누어 계산한 균일성 지표

$$
\frac{\text{Max} - \text{Min}}{2 \times \text{Mean}} \times 100 \%
$$

- 측정값 전체의 범위를 강조
- 최대값과 최소값만 사용하여 계산하므로 극단저인 값(이상치)에 민감
- 전체 데이터의 분포나 편차는 고려하지 않음.
- 직관적이며, 현장에서 쉽게 사용 가능

**1-Sigma 균일성** (Sigma-based Uniformity) :
- 측정값의 통계적 분포를 나타내는 표준편차(1σ)

$$
\frac{\sigma}{\text{Mean}} \times 100 \%
$$

- 측정값의 분포를 통계적으로 평가하여 데이터가 평균값 주변에 얼마나 모여 있는지 확인함.
- 전체 데이터의 분포와 편차를 고려하기 때문에 이상치의 영향이 덜함.
- 더 많은 데이터를 활용하여 전체 분포의 품질을 나타냄.
- 통계적인 접근이므로 공정의 품질 관리(QC)나 분석에 자주 사용됨.


##### 자유도

- 마음대로 정할 수 있는 숫자의 개수
- 통계에서의 자유도)
    - 평균을 먼저 계산하면 데이터 하나는 자동으로 결정됨.
    - 평균이 정해져 있기 때문에, 마지막 숫자는 자유롭게 정할 수 없게 됨. -> 이미 평균으로 결정되어 버린..
    - 그래서 표준편차를 계산할 때는 (데이터 개수 - 1)로 나눠주는 거야.

### Deposition Rate (증착률)

- 웨이퍼 위에 물질이 쌓이는 속도 = 단위 시간당 박막이 증착되는 두께

#### Deposition Rate가 중요한 이유

- Deposition Rate가 높으면 물질이 빠르게 쌓여 효율적인 생산이 가능하지만, 너무 빠르면 두께가 불균일해질 수 있음.
- 반대로 Deposition Rate가 너무 낮으면 공정 시간이 길어져 생산성이 낮아질 수 있음.

- 최적의 증착률을 유지하는 것이 아주 중요
    - 반도체의 품질도 좋고 생산 효율도 좋은 가장 이상적인 속도


#### Deposition Rate 계산 방법

$$
\text{Deposition Rate} \, (\text{nm}/\text{s}) = \frac{\text{Average Thickness} \, (\text{nm})}{\text{Deposition Time} \, (\text{s})}
$$


#### 반도체 공정에서 증착률 활용 방법

1) 품질 관리 및 균일도 조절
2) 생산성 관리 및 비용 절감
3) 제조 공정의 최적화 및 기준 설정

````python
# 파이썬을 이용한 통계적 특성 분석

import pandas as pd

df = pd.read_csv('03_wafer_receipe_1.csv')

df.head()
````

````python
# df에서 박막 두께를 측정한 열만 선택해서 저장

thickness_df = df.loc[:, "THK_Pos_1":"THK_Pos_10"]

thickness_df.head()
````

````python
# thickness_df에 있는 10개의 박막 두께를 행 별로 평균을 내서 Mean_Thickness라는 새로운 열을 만들어서 추가
# axis=1은 가로 방향(행 방향)으로 계산한다는 뜻
df['Mean_Thickness'] = thickness_df.mean(axis=1)

# 최대값은 Max_Thickness 열로 만들어서 추가
df['Max_Thickness'] = thickness_df.max(axis=1)

# 최소값은 Min_Thickness 열로 만들어서 추가
df['Min_Thickness'] = thickness_df.min(axis=1)

# 표준편차를 Std_Thickness 열로 만들어서 추가
df['Std_Thickness'] = thickness_df.std(axis=1)

# 균일도 계산하여 Unif 열로 만들어서 추가
# 균일도 = (최대 두께 - 최소 두께) / (2 x 평균 두께) x 100
df['Unif'] = (df['Max_Thickness'] - df['Min_Thickness']) / (2 * df['Mean_Thickness']) * 100

# Deposition Rate를 계산하여 열 추가
# Deposition Rate = Mean_Thickness / Time
df['Deposition_Rate'] = df['Mean_Thickness'] / df['Time']


df.head()
````

````python
# 새롭게 계산된 데이터를 csv 파일 형태로 저장
# index=False : 데이터 저장 시 별도의 행 번호를 저장하지 않음.
# encoding='utf-8-sig' : 파일의 한글이나 특수 문자가 깨지지 않도록 안전하게 저장하기 위한 옵션

df.to_csv('03_wafer_receipe_2.csv', index=False, encoding='utf-8-sig')
````

### 최적 증착 공정 레시피 탐색

#### 최적의 레시피

반도체 제조 공정에서 목표하는 특성(두께, 균일성)을 가장 효율적으로 달성할 수 있는 공정 변수들의 조합
- 이상적인 레시피는 제품 품질을 극대화하면서 동시에 생산 비용과 시간을 최소화함.

- 원하는 품질을 정확히 만족하면서 효율적이고 경제적인 제조 조건을 찾는 것을 **"최적의 레시피를 찾는다"** 라고 표현함.

#### 최적의 레시피를 찾는 과정

1. 목표 품질 기준 설정
2. 다양한 레시피로 실험 진행
3. 실험 결과 분석
4. 최적의 레시피 선정

**최적의 레시피 선정 효과**
- 제품 품질 향상
- 비용 절감 및 효율성 증가
- 경쟁력 강화

### 최적의 레시피 탐색

증착 공정에서 박막 두께의 목표값이 3100nm이고 허용 범위가 3000 ~ 3200nm일 때, 이 조건을 만족하는 최적의 레시피를 찾아보자.

1. 목표 설정 (목표값과 범위)
- 측정한 10개 지점의 두께가 **정확히 3100nm**인 반도체 박막을 만들고 싶음.
- 약간의 오차 포함 **3000 ~ 3200nm**를 허용 범위로 정함.

2. 여러 레시피로 실험하기
- 사용하는 가스 양 (Gas_Si2H6)
- 공정에서 사용하는 전력 (Power)
- 공정 시 온도 (Temperature)
- 공정 시간 (Time)

3. 조건에 맞는 레시피 추려내기
- 박막의 가장 얇은 부분(최소값) $\ge$ 3000nm
- 박막의 가장 두꺼운 부분(최대값) $\le$ 3200nm

4. 가장 좋은 레시피의 조건 선택하기
- 목표값에 얼마나 가까운가?
- 균일성이 얼마나 뛰어난가?
- 표본 표준편차가 얼마나 낮은가?
- 증착률이 높은가?

1) 두께 점수 계산
2) 균일성 점수 계산
3) 표준편차 점수 계산
4) 증착 속도 점수 계산

5. 최종 점수 계산
- 네 가지 점수에 각각의 가중치를 곱해 최종 점수를 계산

```scss
최종 점수 = (0.4 x 두께 점수) + (0.2 x 균일성 점수) + (0.2 x 표준편차 점수) + (0.2 x 증착 속도 점수)
```

- 최종 점수가 가장 높은 레시피가 **최적의 레시피 !**

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')

df = pd.read_csv('03_wafer_receipe_2.csv')

df.head()
````

````python
# 목표로 하는 박막의 두께 = 3100nm
target_thickness = 3100

# 허용 오차 (± 100nm)
tolerance = 100

# 여러 레시피 중 목표 범위(3100 ± 100nm)에 들어가는 레시피만 뽑아내자.

# 데이터 프레임 조건 탐색
# 최소 두께가 3000 이상 최대 두께가 3200 이하인 조건을 이용해 필터링

filtered_recipes = df[
    (df['Min_Thickness'] >= target_thickness - tolerance) &  # 최소 두께가 3000nm 이상이고
    (df['Max_Thickness'] <= target_thickness + tolerance)    # 최대 두께가 3200nm이하인 조건.
].copy()  # 데이터를 안전하게 독립적으로 관리할 수 있도록 복사본 만들기 (원본 훼손 방지)

filtered_recipes
````

````python
# 목표 두께와 각 레시피 평균 두께와의 차이 중 가장 큰 값 찾기
# abs() : 절대값 반환

max_thickness_diff = abs(filtered_recipes['Mean_Thickness'] - 3100).max()

print(f"(평균 두께-목표두께) 중 가장 큰 값: {max_thickness_diff}")


# 각 레시피가 목표 두께에 얼마나 가까운지 평가하는 점수 계산
# 목표에 가까울수록 점수가 1에 가까움.
# 두께 점수 = 1 - ((|평균두께 - 목표두께|) / 모든 레시피 중 목표와 가장 큰 두께 차이)

filtered_recipes['Score_Thickness'] = 1 - (abs(filtered_recipes['Mean_Thickness'] - 3100) / max_thickness_diff)

filtered_recipes.head()
````

````python
# 균일성의 최대값과 최소값의 차이

max_unif_diff = filtered_recipes['Unif'].max() - filtered_recipes['Unif'].min()

print(f"가장 나쁜 균일성 값: {filtered_recipes['Unif'].max():.2f}")
print(f"가장 좋은 균일성 값: {filtered_recipes['Unif'].min():.2f}")
print(f"가장 나쁜 균일성 값-가장 좋은 균일성 값: {max_unif_diff}")


# 균일성 점수 계산
# 균일성이 좋을수록(값이 작을수록) 점수가 1에 가까움.
# 균일성 점수 = 1 - ((해당 레시피 균일성 값 - 가장 좋은 균일성 값) / (가장 나쁜 균일성 값 - 가장 좋은 균일성 값))

filtered_recipes['Score_Unif'] = 1 - ((filtered_recipes['Unif'] - filtered_recipes['Unif'].min()) / max_unif_diff)

filtered_recipes.head()
````

````python
# 박막 두께의 표준편차의 최대값과 최소값 차이 구하기

max_std_diff = filtered_recipes['Std_Thickness'].max() - filtered_recipes['Std_Thickness'].min()

print(f"가장 높은 표준편차: {filtered_recipes['Std_Thickness'].max():.2f}")
print(f"가장 낮은 표준편차: {filtered_recipes['Std_Thickness'].min():.2f}")
print(f"가장 높은 표준편차-가장 낮은 표준편차: {max_std_diff}")


# 표준편차 점수 계산
# 표준편차가 작을수록(박막이 균일할수록) 점수가 1에 가까움.
# 표준편차 점수 = 1 - ((해당 레시피 표준편차 값 - 가장 낮은 표준편차) / (가장 높은 표준편차 - 가장 낮은 표준편차))

filtered_recipes['Score_Std'] = 1 - ((filtered_recipes['Std_Thickness'] - filtered_recipes['Std_Thickness'].min()) / max_std_diff)

filtered_recipes.head()
````

````python
# 증착속도의 최대값과 최소값 차이 구하기

max_deposition_diff = filtered_recipes['Deposition_Rate'].max() - filtered_recipes['Deposition_Rate'].min()

print(f"가장 높은 속도: {filtered_recipes['Deposition_Rate'].max():.2f}")
print(f"가장 낮은 속도: {filtered_recipes['Deposition_Rate'].min():.2f}")
print(f"가장 높은 속도-가장 낮은 속도: {max_deposition_diff}")


# 증착속도 점수 계산
# 증착속도가 빠를수록 점수가 1에 가까움.
# 증착속도 점수 = (해당 레시피 속도 - 가장 낮은 속도) / (가장 높은 속도 - 가장 낮은 속도)

filtered_recipes['Score_Deposition'] = (filtered_recipes['Deposition_Rate'] - filtered_recipes['Deposition_Rate'].min()) / max_deposition_diff

filtered_recipes.head()
````

````python
# 평가 지표별로 중요도(가중치) 설정

weight_thickness = 0.4  # 두께가 목표값과 가까운 것이 가장 중요 (40%)
weight_unif = 0.2  # 균일성 중요도 (20%)
weight_std = 0.2  # 두께 표준편차의 중요도 (20%)
weight_deposition = 0.2  # 증착속도 중요도 (20%)


# 최종 점수를 계산하여 'Final_Score' 열에 저장
filtered_recipes['Final_Score'] = (
    filtered_recipes['Score_Thickness'] * weight_thickness +
    filtered_recipes['Score_Unif'] * weight_unif +
    filtered_recipes['Score_Std'] * weight_std +
    filtered_recipes['Score_Deposition'] * weight_deposition
)


# 최종 점수로 정렬하여 최적의 레시피 찾기
filtered_recipes = filtered_recipes.sort_values(by='Final_Score', ascending=False)

filtered_recipes.head()
````

````python
# iloc[0] : 정렬된 데이터에서 가장 첫 번째 행(최고점)을 선택

optimal_recipe = filtered_recipes.iloc[0]

print("✅ 최적의 레시피 정보 ✅")
print(optimal_recipe[['Recipe_Name', 'Mean_Thickness', 'Min_Thickness', 'Max_Thickness', 'Unif', 'Std_Thickness', 'Deposition_Rate', 'Final_Score']])
````
